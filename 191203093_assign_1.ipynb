```python
#Import Command
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#Reading CSV Data file 
data = pd.read_csv("https://ftp.cs.wisc.edu/math-prog/cpo-dataset/machine-learn/cancer/WDBC/WDBC.dat")
```

```python
data.head()
```

```python
data.tail()
```

```python
#Getting the useful insights about the data
data.describe()
```

```python
#Checking the target variable count
data['diagnosis'].value_counts()
```

```python
data.head()
```

```python
#Checking the co-relation of the entire dataset
sns.set(rc = {'figure.figsize':(20,20)})
sns.heatmap(data.corr(), annot = True, fmt='.2g',cmap= 'coolwarm')
```

```python
#Dropping the least important features
data = data.drop(labels=['id','Unnamed: 32'],axis=1)
```

```python
data.head()
```

```python
#Checking if there is any missing Data in the Dataset
data.isnull().sum()
```

```python
#As the diagnosis is the target ie the dependent feature and it is in the categorical form
#we have to assign class numbers to them
#Here we can consider 0:B and 1:M
from sklearn.preprocessing import LabelEncoder
labelencoder = LabelEncoder()
data['Target'] = labelencoder.fit_transform(data['diagnosis'])
```

```python
data.head()
```

```python
data.tail()
```

```python
data = data.drop(labels=['diagnosis'],axis=1)
```

```python
data.head()
```

Separating the Featues and Target Variables

```python
Data_x = data.drop(labels=['Target'],axis=1)#Collection of all the Independent Features
Data_y = data['Target']#collecton of all the target/dependent Variables
```

```python
#Separating the Test and train data
from sklearn.model_selection import train_test_split
train_x,test_x,train_y,test_y = train_test_split(Data_x,Data_y,test_size=0.25,random_state=42)
```

```python
#Using Accuracy Score to predict the Accuracy of Each Model
from sklearn.metrics import accuracy_score
```

```python
#Using the KNN Model
from sklearn.neighbors import KNeighborsClassifier
KNN = KNeighborsClassifier()
```

```python
#Train the Model
KNN.fit(train_x,train_y)
```

```python
#Predicting for New Values
Predict1 = KNN.predict(test_x)
```

```python
#Checking the Accuracy of the Model
Acc1 = accuracy_score(test_y,Predict1)*100
print("Accuracy of the KNN Model : ",Acc1,"%")
```

```python
#Using the Support Vector Machine Classifier
from sklearn.svm import SVC
SVM = SVC()
```

```python
#training the SVM classifier
SVM.fit(train_x,train_y)
```

```python
#predicting using the Test Data
Predict2 = SVM.predict(test_x)
```

```python
#Checking the Accuracy of the Model
Acc2 = accuracy_score(test_y,Predict2)*100
print("Accuracy of the KNN Model : ",Acc2,"%")
```

```python
#Using the Decision Tree Classifier
from sklearn.tree import DecisionTreeClassifier
DTC = DecisionTreeClassifier()
```

```python
#Train the Model
DTC.fit(train_x,train_y)
```

```python
#Predicting using the Test Data
Predict3 = DTC.predict(test_x)
```

```python
#Checking the Accuracy of the Model
Acc3 = accuracy_score(test_y,Predict3)*100
print("Accuracy of the KNN Model : ",Acc3,"%")
```

```python
Accuracy = [Acc1,Acc2,Acc3]
plt.plot(Accuracy)
```
